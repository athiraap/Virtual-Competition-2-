```python
import pandas as pd
import matplotlib.pyplot as plt  # Don't forget to import matplotlib

# Load the dataset
dataset_path = '/content/drive/MyDrive/DSA/Train.csv'
train_data = pd.read_csv(dataset_path)

# Plot bar chart for 'Gender'
train_data['Gender'].value_counts(normalize=True).plot.bar(title='Gender', color='c')
plt.show()

# Plot bar chart for 'Married'
train_data['Married'].value_counts(normalize=True).plot.bar(title='Married', color='c')
plt.show()

# Plot bar chart for 'Self_Employed'
train_data['Self_Employed'].value_counts(normalize=True).plot.bar(title='Self_Employed', color='c')
plt.show()

# Plot bar chart for 'Credit_History'
train_data['Credit_History'].value_counts(normalize=True).plot.bar(title='Credit_History', color='c')
plt.show()
```

```python
import pandas as pd

# Assuming train_data is already loaded
# Load the dataset
dataset_path = '/content/drive/MyDrive/DSA/Train.csv'
train_data = pd.read_csv(dataset_path)

# Drop the 'Loan_ID' column
train_data.drop(['Loan_ID'], axis=1, inplace=True)

# Display the modified DataFrame
print(train_data.head())
```
